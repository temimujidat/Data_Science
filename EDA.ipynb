```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#imports the data set

data_set = pd.read_csv('train.csv')
```

```python
#Understanding the data set - This gives the first 5 rows

data_set.head()
```

```python
#Understanding the data set - This tells how many rows and columns

data_set.shape
```

```python
#this shows the attributes that have missing values and the total number of missing values present

data_set.isnull().sum()
```

```python
#This shows a clear graphical view of the attributes that has missing values

sns.heatmap(data_set.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
#Understanding the data set - This gives the unique variables in the Pclass attribute

data_set['Pclass'].unique()
```

```python
#cleaning the data set ---- Removing unwanted attributes

data_set.drop(['Name','Ticket','Cabin'],axis=1,inplace=True)
```

```python
data_set.shape
```

```python
data_set.head()
```

```python
#A clear graphical view of the relationship between two variables

sns.pairplot(data_set)
```

```python
#clear graphical view of the relationship between 2 numerical variables
# This shows us that more females survived than males. And most of these females are from ages 12 and above

sns.relplot(x='Survived', y='Age', hue='Sex', data=data_set)
```

```python
#This shows the information about 1 variable.
#We can denote that we have more of people that did not survive than people that survived

sns.countplot(x='Survived', data = data_set)
```

```python
sns.distplot(data_set['Survived'],kde=False, color='darkred',bins=10)
```

```python
#graphical relationship between 2 variables
#We can denote that more females survived than male

sns.countplot(x='Survived', hue='Sex', data=data_set,palette='rainbow')
```

```python
#graphical relationship between 2 variables
#We can denote that the highest number of passengers that did not survive are from the Pclass 3.. i.e 3rd class

sns.countplot(x='Survived', hue='Pclass', data=data_set, palette='rainbow')
```

```python
#shows information about the Age attribute without considering the NA values
#We can denote that most passengers are between ages 18 to 36

sns.distplot(data_set['Age'].dropna(),kde=False,color='darkred',bins=40)
```

```python
#This is a box plot that shows us the average age value of passengers in each of the Pclass. E.g Pclass 1 has average age value of 38

sns.boxplot(x='Pclass',y='Age',data=data_set,palette='winter')
```
