```python
import pandas as pd
```

```python
data2= pd.read_csv('african_crises.csv')
```

```python
data2.head()
```

```python
country_list = data2['country'].unique()

for i,country in enumerate(country_list):
#     print(i)
#     print(country)
    data2.replace(country,i,inplace=True)
```

```python
cc3_list = data2['cc3'].unique()

for i,cc3 in enumerate(cc3_list):
#     print(i)
#     print(cc3)
    data2.replace(cc3,i,inplace=True)
```

```python
banking_crisis_list = data2['banking_crisis'].unique()

for i,banking_crisis in enumerate(banking_crisis_list):
#     print(i)
#     print(banking_crisis)
    data2.replace(banking_crisis,i,inplace=True)
```

```python
data2.corr(method="pearson")
```

```python
This analysis shows that there is a STRONG correlation relationship between the following :
    - 99% correlation between "case" and "cc3"
    - 99% correlation between "case" and "country"
    - 100% correlation between "cc3" and "country"
    
This analysis shows that there is a MEDIUM correlation relationship between the following :
    - 41% correlation between "independence" and "year"
    - 42% correlation between "sovereign_external_debt_default" and "exch_usd"
    
SUMMARY::::::::::::
    - Since the attributes 'cc3', 'case' and 'country' relate strongly, we can drop any two of the three attributes becase ONE is enough information
    
```
