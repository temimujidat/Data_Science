```python
#plotting a single graph that shows the yearly growth in exchange rate in each africa country
#it is concluded that "Central African Republic" has the heighest exchange rate.


africa_data=['egypt_data', 'SA_data', 'zimbabwe_data', 'algeria_data', 'angola_data', 'tunisia_data', 'morocco_data', 'zambia_data', 'mauritius_data', 'kenya_data', 'ivorycoast_data', 'nigeria_data', 'centralafricanrepublic_data']
plt.figure(figsize=(10,6))

plt.plot(egypt_data.year, egypt_data.exch_usd)
plt.plot(SA_data.year, SA_data.exch_usd)
plt.plot(zimbabwe_data.year, zimbabwe_data.exch_usd)
plt.plot(algeria_data.year, algeria_data.exch_usd)
plt.plot(angola_data.year, angola_data.exch_usd)
plt.plot(tunisia_data.year, tunisia_data.exch_usd)
plt.plot(morocco_data.year, morocco_data.exch_usd)
plt.plot(zambia_data.year, zambia_data.exch_usd)
plt.plot(mauritius_data.year, mauritius_data.exch_usd)
plt.plot(kenya_data.year, kenya_data.exch_usd)
plt.plot(ivorycoast_data.year, ivorycoast_data.exch_usd)
plt.plot(nigeria_data.year, nigeria_data.exch_usd)
plt.plot(centralafricanrepublic_data.year, centralafricanrepublic_data.exch_usd)
plt.legend(africa_data)
plt.xlabel('year')
plt.ylabel('exch_usd')
plt.show()
```

```python
#this relationship graph between banking_crisis and systemic_crisis

#The graph shows that the banking_crisis is directly proportional to the systemic_crisis
# In other words, the higher the banking crisis, the lower the systematic crisis. and vice versa

#The next cell below show the correlation relationship between the two(Banking_crisis and systematic crisis)


sns.relplot(x='banking_crisis', y='systemic_crisis', data=crisis_data)
```

```python
#This helps to convert the banking_crisis attributes to numeriacl attributes so to be included in the correlation relationship table

banking_crisis_list = crisis_data['banking_crisis'].unique()

for i,banking_crisis in enumerate(banking_crisis_list):
       #print(i)
      #print(banking_crisis)
      crisis_data.replace(banking_crisis,i,inplace=True)
    
```

```python
#correlation relationship of the data set

#The corelation relationship shows that there is a strong negative correlation between "banking_crisis" and "Systematic_crisis"
#The value of the strong negative correlation is: -0.854 
#The correlation value signifies that as tha banking_crisis in the African Economy decreases, the systematic_crisis will increase. and vice versa

crisis_data.corr(method="pearson")
```
