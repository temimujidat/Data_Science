```python
#plotting a single graph that shows the yearly growth in exchange rate in each africa country
#it is concluded that "Central African Republic" has the heighest exchange rate.


africa_data=['egypt_data', 'SA_data', 'zimbabwe_data', 'algeria_data', 'angola_data', 'tunisia_data', 'morocco_data', 'zambia_data', 'mauritius_data', 'kenya_data', 'ivorycoast_data', 'nigeria_data', 'centralafricanrepublic_data']
plt.plot(egypt_data.year, egypt_data.exch_usd)
plt.plot(SA_data.year, SA_data.exch_usd)
plt.plot(zimbabwe_data.year, zimbabwe_data.exch_usd)
plt.plot(algeria_data.year, algeria_data.exch_usd)
plt.plot(angola_data.year, angola_data.exch_usd)
plt.plot(tunisia_data.year, tunisia_data.exch_usd)
plt.plot(morocco_data.year, morocco_data.exch_usd)
plt.plot(zambia_data.year, zambia_data.exch_usd)
plt.plot(mauritius_data.year, mauritius_data.exch_usd)
plt.plot(kenya_data.year, kenya_data.exch_usd)
plt.plot(ivorycoast_data.year, ivorycoast_data.exch_usd)
plt.plot(nigeria_data.year, nigeria_data.exch_usd)
plt.plot(centralafricanrepublic_data.year, centralafricanrepublic_data.exch_usd)
plt.legend(africa_data)
plt.xlabel('year')
plt.ylabel('exch_usd')
plt.show()
```

```python
#this representation reveals that about 15% of african countries have a sovereign_external_debt_default within the year of observation. while 85% of the african countries have no sovereign_external_debt_default

sns.countplot(x='sovereign_external_debt_default', data = crisis_data)
```

```python
#this representation shows that "Egypt" remains the african country with the least sovereign_external_debt_default within the complete year of study
#Also, this representation shows that "Central African Republic" remains the african country with the most sovereign_external_debt_default within the complete year of study

sns.countplot(x='sovereign_external_debt_default', hue='country', data=crisis_data,palette='rainbow')
```
