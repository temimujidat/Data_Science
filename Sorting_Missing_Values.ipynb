```python
import pandas as pd
from sklearn.impute import SimpleImputer
import numpy as np
```

```python
data = pd.read_csv('adult.csv', names = titles)
```

```python
titles=["age","workplace","fnlwgt","education","education-num","marital-status","occupation","relationship","race","sex","capital-gain","capital-loss","hours-per-week","native-country","amount"]

```

```python
new_data=data.replace(' ?', np.NaN)
new_data[200:300]
```

```python
numpy_data = new_data.to_numpy()
```

```python
imp = SimpleImputer(missing_values=np.NaN, strategy='most_frequent')
```

```python
clean_data = imp.fit_transform(numpy_data)
```

```python
clean_dataframe = pd.DataFrame(data=clean_data,columns=titles)
```

```python
clean_dataframe[200:300]
```

```python
clean_dataframe.to_csv('clean_data.csv')
```
